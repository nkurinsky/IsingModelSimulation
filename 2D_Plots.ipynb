```python
import numpy
from pylab import *
%matplotlib inline
```

```python
data=numpy.loadtxt("2D_results.txt")
dim=data[:,0]
size=data[:,1]
q=data[:,2]
temp=data[:,3]
mag=data[:,4]
emag=data[:,5]
c_inds=numpy.arange(6,16,2)
corr=data[:,c_inds]

figure(figsize=(8,6))
sizes=numpy.unique(size)
pts=numpy.where(q == 2)
ising_temp=temp[pts]
ising_mag=mag[pts]
ising_corr=corr[pts]
ising_size=size[pts]
for i in range(0,len(sizes)):
    inds=where(ising_size==sizes[i])
    plot(ising_temp[inds],ising_mag[inds],label=str(sizes[i]))
legend(loc="upper right",title="Size")
xlim(0.5,2.5)
ylim(0,1.2)
show()

clf()
figure(figsize=(8,6))
qs=numpy.unique(q)
pts=numpy.where(size == 100)
l_temp=temp[pts]
l_mag=mag[pts]
l_corr=corr[pts]
l_q=q[pts]
for i in range(0,len(qs)):
    inds=where(l_q==qs[i])
    plot(l_temp[inds],l_mag[inds],label=str(qs[i]))
legend(loc="upper right",title="Q")
xlim(0.5,2)
ylim(0,1.2)
show()

figure(figsize=(12,6))
lcorr1=l_corr[:,0]
lcorr2=l_corr[:,2]
lcorr3=l_corr[:,4]
print(len(lcorr1))
for i in range(0,len(qs)):
    if(i == 0):
        color="black"
    if(i == 1):
        color="blue"
    if(i == 2):
        color="green"
    if(i == 3):
        color="red"
    if(i == 4):
        color="orange"
    inds=where(l_q==qs[i])
    plot(l_temp[inds],lcorr1[inds],label="r1, q="+str(qs[i]),color=color)
    plot(l_temp[inds],lcorr2[inds],'--',label="r2, q="+str(qs[i]),color=color)
    plot(l_temp[inds],lcorr3[inds],':',label="r3, q="+str(qs[i]),color=color)
legend(loc="center right",title="r, Q")
xlim(0.5,3)
ylim(0,1.2)
show()
```
