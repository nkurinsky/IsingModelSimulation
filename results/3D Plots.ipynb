```python
import numpy
from pylab import *
import scipy
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
data=numpy.loadtxt("3cD_results.txt")
dim=data[:,0]
size=data[:,1]
sz=data[:,1]
q=data[:,2]
temp=data[:,3]
mag=data[:,4]
emag=data[:,5]
c_inds=numpy.arange(6,26,2)
corr=data[:,c_inds]
ecorr=data[:,c_inds+1]

gpts=numpy.where(q < 100)
dim=dim[gpts]
size=size[gpts]
sz=sz[gpts]
q=q[gpts]
temp=temp[gpts]
mag=mag[gpts]
emag=emag[gpts]
corr=corr[gpts]
ecorr=ecorr[gpts]

qs=numpy.unique(q)
print(qs)
for i in range(0,len(qs)):
    figure(figsize=(8,6))
    sizes=numpy.unique(size)
    pts=numpy.where(q == qs[i])
    ising_temp=temp[pts]
    ising_mag=mag[pts]
    ising_corr=corr[pts]
    ising_size=size[pts]
    for j in range(0,len(sizes)):
        inds=where(ising_size==sizes[j])
        plot(ising_temp[inds],ising_mag[inds],label=str(sizes[j]))
    legend(loc="upper right",title="Size")
    xlim(0,5.0)
    ylim(0,1.2)
    savefig("q"+str(qs[i])+"_size_effect_3d.png")
```

```python
clf()
figure(figsize=(12,6))
qs=numpy.unique(q)
pts=numpy.where(size == 100)
l_temp=temp[pts]
l_mag=mag[pts]
l_corr=corr[pts]
l_ecorr=ecorr[pts]
l_q=q[pts]
for i in range(0,len(qs)):
    inds=where(l_q==qs[i])
    if(i == 0):
        color="black"
    if(i == 1):
        color="blue"
    if(i == 2):
        color="green"
    if(i == 3):
        color="red"
    if(i == 4):
        color="orange"
    plot(l_temp[inds],l_mag[inds],label=str(qs[i]),color=color)
legend(loc="upper right",title="Q")
xlim(1.0,2.5)
ylim(0,1.2)
show()

figure(figsize=(12,6))
lcorr1=l_corr[:,0]
lcorr2=l_corr[:,2]
lcorr3=l_corr[:,4]
for i in range(0,len(qs)):
    if(i == 0):
        color="black"
    if(i == 1):
        color="blue"
    if(i == 2):
        color="green"
    if(i == 3):
        color="red"
    if(i == 4):
        color="orange"
    inds=where(l_q==qs[i])
    plot(l_temp[inds],lcorr1[inds],label="r1, q="+str(qs[i]),color=color)
    plot(l_temp[inds],lcorr2[inds],'--',label="r2, q="+str(qs[i]),color=color)
    plot(l_temp[inds],lcorr3[inds],':',label="r3, q="+str(qs[i]),color=color)
legend(loc="center right",title="r, Q")
xlim(0,4)
ylim(0,1.2)
show()
```

```python
def func(r,a):
    return -a*r-1.063*log(r)

rlim=7
figure(figsize=(10,6))
x=numpy.arange(1,rlim,1)
fakezero=1e-4
Tcr=qs*0.0
for i in range(0,len(qs)):
    if(i == 0):
        color="black"
    if(i == 1):
        color="blue"
    if(i == 2):
        color="green"
    if(i == 3):
        color="red"
    if(i == 4):
        color="orange"
    inds=where(l_q==qs[i])

    xtemp=l_temp[inds]
    xcorr=l_corr[inds]
    xmag=l_mag[inds]
    
    xi=xtemp*0.0
    for j in range(0,len(xcorr)):
        icorr=xcorr[j][1:rlim]
        pts=numpy.where(icorr < fakezero)
        icorr[pts]=fakezero
        popt,pcov=curve_fit(func,x,log(icorr),p0=(1e-5))
        xi[j]=1/popt[0]
   
    Tcr[i]=xtemp[numpy.argmax(xi)]
    plot(xtemp,xi,label="Corr(r), Tc="+str(Tcr[i]),color=color)

xlabel("Temperature")
ylabel("Relaxation Length")
ylim(0,5)
legend(loc="upper right")

def func2(r,a,b,c):
    return a*exp(-r/c)/(pow(r,b))

figure(figsize=(10,6))
x=numpy.arange(1,rlim,1)
fakezero=1e-10
for i in range(0,len(qs)):
    if(i == 0):
        color="black"
    if(i == 1):
        color="blue"
    if(i == 2):
        color="green"
    if(i == 3):
        color="red"
    if(i == 4):
        color="orange"
    inds=where(l_q==qs[i])

    xtemp=l_temp[inds]
    xcorr=l_corr[inds]
    excorr=l_ecorr[inds]
    xmag=l_mag[inds]
    
    xi=xtemp*0.0
    gpts=where(xtemp==Tcr[i])
    icorr=xcorr[gpts[0][0]][1:rlim]
    eicorr=excorr[gpts[0][0]][1:rlim]
    pts=numpy.where(icorr < fakezero)
    icorr[pts]=fakezero
    plot(x,icorr,':',color=color)
    popt,pcov =curve_fit(func2,x,icorr,sigma=eicorr,p0=(1.0,0.25,10.0))
    print(popt)
    print(pcov)
    #plot(x,icorr,':',color=color)
    plot(x,func2(x,popt[0],popt[1],popt[2]),label="Corr(r), Tc="+str(Tcr[i])+", Nu="+"{0:5.3f}".format(popt[1]),color=color)
    
xlabel("Temperature")
ylabel("Relaxation Length")
legend(loc="upper right")
```

```python
def func(r,a):
    return -a*r-0.9*log(r)

def func2(r,a,b,c):
    return a*exp(-r/c)/(pow(r,b))

sizes=numpy.unique(size)
gpts=numpy.where(qs != 4)
qsq=qs[gpts]
for qi in range(0,len(qsq)):
    figure(figsize=(8,6))
    pts=numpy.where(q == qsq[qi])
    l_temp=temp[pts]
    l_mag=mag[pts]
    l_corr=corr[pts]
    l_ecorr=ecorr[pts]
    lsize=sz[pts]

    figure(figsize=(10,6))
    x=numpy.arange(1,rlim,1)
    fakezero=1e-4
    Tcr=sizes*0.0
    for i in range(1,len(sizes)):
        if(i == 0):
            color="magenta"
        if(i == 1):
            color="blue"
        if(i == 2):
            color="green"
        if(i == 3):
            color="red"
        if(i == 4):
            color="orange"
        if(i == 5):
            color="black"
        inds=where(lsize==sizes[i])

        xtemp=l_temp[inds]
        xcorr=l_corr[inds]
        xmag=l_mag[inds]
    
        xi=xtemp*0.0
        for j in range(0,len(xcorr)):
            icorr=xcorr[j][1:rlim]
            pts=numpy.where(icorr < fakezero)
            icorr[pts]=fakezero
            popt,pcov=curve_fit(func,x,log(icorr),p0=(1e-5))
            xi[j]=1/popt[0]
   
        Tcr[i]=xtemp[numpy.argmax(xi)]
        plot(xtemp,xi,label=str(sizes[i])+", Tc="+str(Tcr[i]),color=color)

    xlabel("Temperature")
    ylabel("Relaxation Length")
    ylim(0,5)
    legend(loc="upper right")
```
