```python
import numpy
from pylab import *
import scipy
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
data=numpy.loadtxt("4cD_results.txt")
dim=data[:,0]
size=data[:,1]
sz=data[:,1]
q=data[:,2]
temp=data[:,3]
mag=data[:,4]
emag=data[:,5]
c_inds=numpy.arange(6,26,2)
corr=data[:,c_inds]
ecorr=data[:,c_inds+1]

gpts=numpy.where(size != 100)
dim=dim[gpts]
size=size[gpts]
sz=sz[gpts]
q=q[gpts]
temp=temp[gpts]
mag=mag[gpts]
emag=emag[gpts]
corr=corr[gpts]
ecorr=ecorr[gpts]

qs=numpy.unique(q)
print(qs)
for i in range(0,len(qs)):
    figure(figsize=(8,6))
    sizes=numpy.unique(size)
    pts=numpy.where(q == qs[i])
    ising_temp=temp[pts]
    ising_mag=mag[pts]
    ising_corr=corr[pts]
    ising_size=size[pts]
    for j in range(0,len(sizes)):
        inds=where(ising_size==sizes[j])
        plot(ising_temp[inds],ising_mag[inds],label=str(sizes[j]))
    legend(loc="upper right",title="Size")
    xlim(0,5.0)
    ylim(0,1.2)
    title("Spin Q="+str(int(qs[i])))
    xlabel("Potts Temperature")
    ylabel("Order Parameter Value")
    savefig("q"+str(int(qs[i]))+"_size_effect_4d.png")
```

```python
def func(r,a):
    return -a*r-2.0*log(r)

rlim=8
figure(figsize=(10,6))
x=numpy.arange(1,rlim,1)
fakezero=1e-4
Tcr=qs*0.0
for i in range(0,len(qs)):
    if(i == 0):
        color="black"
    if(i == 1):
        color="blue"
    if(i == 2):
        color="green"
    if(i == 3):
        color="red"
    if(i == 4):
        color="orange"
    inds=where(l_q==qs[i])

    xtemp=l_temp[inds]
    xcorr=l_corr[inds]
    xmag=l_mag[inds]
    
    xi=xtemp*0.0
    for j in range(0,len(xcorr)):
        icorr=xcorr[j][1:rlim]
        pts=numpy.where(icorr < fakezero)
        icorr[pts]=fakezero
        popt,pcov=curve_fit(func,x,log(icorr),p0=(1e-5))
        xi[j]=1/popt[0]
   
    Tcr[i]=xtemp[numpy.argmax(xi)]
    plot(xtemp,xi,label="Corr(r), Tc="+str(Tcr[i]),color=color)

xlabel("Temperature")
ylabel("Relaxation Length")
ylim(0,3)
legend(loc="upper right")

def func2(r,a,b,c):
    return a*exp(-r/c)/(pow(r,b))

figure(figsize=(10,6))
x=numpy.arange(1,rlim,1)
fakezero=1e-10
for i in range(0,len(qs)):
    if(i == 0):
        color="black"
    if(i == 1):
        color="blue"
    if(i == 2):
        color="green"
    if(i == 3):
        color="red"
    if(i == 4):
        color="orange"
    inds=where(l_q==qs[i])

    xtemp=l_temp[inds]
    xcorr=l_corr[inds]
    excorr=l_ecorr[inds]
    xmag=l_mag[inds]
    
    xi=xtemp*0.0
    gpts=where(xtemp==Tcr[i])
    icorr=xcorr[gpts[0][0]][1:rlim]
    eicorr=excorr[gpts[0][0]][1:rlim]
    pts=numpy.where(icorr < fakezero)
    icorr[pts]=fakezero
    plot(x,icorr,':',color=color)
    popt,pcov =curve_fit(func2,x,icorr,sigma=eicorr,p0=(1.0,0.25,10.0))
    print(popt)
    #plot(x,icorr,':',color=color)
    plot(x,func2(x,popt[0],popt[1],popt[2]),label="Corr(r), Tc="+str(Tcr[i])+", Nu="+"{0:5.3f}".format(popt[1]),color=color)
    
xlabel("Temperature")
ylabel("Relaxation Length")
legend(loc="upper right")
```
