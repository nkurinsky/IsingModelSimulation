```python
import numpy
from pylab import *
%matplotlib inline
```

```python
data=numpy.loadtxt("results.txt")
dim=data[:,0]
size=data[:,1]
q=data[:,2]
temp=data[:,3]
mag=data[:,4]
emag=data[:,5]
c_inds=numpy.arange(6,16,2)
corr=data[:,c_inds]

figure(figsize=(8,6))
sizes=numpy.unique(size)
pts=numpy.where(q == 2)
ising_temp=temp[pts]
ising_mag=mag[pts]
ising_corr=corr[pts]
ising_size=size[pts]
for i in range(0,len(sizes)):
    inds=where(ising_size==sizes[i])
    plot(ising_temp[inds],ising_mag[inds],label=str(sizes[i]))
legend(loc="upper right",title="Size")
xlim(0.5,3)
ylim(0,1.2)
show()

clf()
figure(figsize=(8,6))
qs=numpy.unique(q)
pts=numpy.where(size == 100)
l_temp=temp[pts]
l_mag=mag[pts]
l_corr=corr[pts]
l_q=q[pts]
for i in range(0,len(qs)):
    inds=where(l_q==qs[i])
    plot(l_temp[inds],l_mag[inds],label=str(qs[i]))
legend(loc="upper right",title="Q")
xlim(0.5,2)
ylim(0,1.2)
show()

figure(figsize=(8,6))
lcorr1=l_corr[:,0]
lcorr2=l_corr[:,2]
lcorr3=l_corr[:,4]
print(len(lcorr1))
for i in range(0,len(qs)):
    if(i == 0):
        color="black"
    if(i == 1):
        color="blue"
    if(i == 2):
        color="green"
    if(i == 3):
        color="red"
    inds=where(l_q==qs[i])
    plot(l_temp[inds],lcorr1[inds],label="r1, q="+str(qs[i]),color=color)
    plot(l_temp[inds],lcorr2[inds],'--',label="r2, q="+str(qs[i]),color=color)
    plot(l_temp[inds],lcorr3[inds],':',label="r3, q="+str(qs[i]),color=color)
legend(loc="upper left",title="Q")
xlim(0.5,2)
ylim(0,1.2)
show()
```

```python
data=numpy.loadtxt("results.txt",skiprows=1)

temp=data[:,0]
mag=data[:,1]
c1=data[:,3]
c2=data[:,5]
c3=data[:,7]
figure(figsize=(10,6))
plot(temp,mag,label=("magnetization"));
plot(temp,c1,label=("r=1 covariance"));
plot(temp,c2,label=("r=2 covariance"));
plot(temp,c3,label=("r=3 covariance"));
ylim(0,1.2)
xlabel("Potts Temperature")
ylabel("Value")
legend(loc="center left")

figure(figsize=(10,6))
inds=numpy.arange(3,11,2)
corr=data[:,inds]
x=numpy.arange(1,len(corr[0])+1,1)
fakezero=1e-4
#for i in range(0,len(corr),10):
    #plot(x,-log(abs(corr[i])+fakezero)+log(1-pow(mag[i],2)+fakezero))
#ylim(0,5);

figure(figsize=(10,6))
gpts=numpy.where(temp == 0.6)
rl=temp[gpts[0][0]:]*0.0
xi=temp[gpts[0][0]:]*0.0
corr=corr[gpts[0][0]:]
temp=temp[gpts[0][0]:]
mag=mag[gpts[0][0]:]
for i in range(0,len(corr)):
    pts=numpy.where(corr[i] < fakezero)
    corr[i][pts]=fakezero
    zeropt=corr[i][0]
    if zeropt<fakezero:
        zeropt=fakezero
    tcorr=-log(corr[i]+mag[i]**2)+log(zeropt)
    mt=mean(tcorr[1:]/x[1:])
    rl[i]=abs(1/mt)
    tcorr=-log(corr[i])+log(zeropt)
    mt=mean(tcorr[1:]/x[1:])
    xi[i]=1/mt
    #xi[i]=1/(tcorr[len(tcorr)-1]/len(tcorr))

xind=numpy.argmax(xi)
rind=numpy.argmax(rl)
Tcx=temp[numpy.argmax(xi)]
Xxi=xi[numpy.argmax(xi)]
Tcr=temp[numpy.argmax(rl)]
Xrl=rl[numpy.argmax(rl)]    
    
plot(temp,rl,label="Corr(r)/Cov(1), Tc="+str(Tcr))
plot(temp,xi,label="Cov(r), Tc="+str(Tcx))
#xlim(1,2.5)
ylim(0,6)
xlabel("Temperature")
ylabel("Relaxation Length")
legend(loc="upper right")
```
